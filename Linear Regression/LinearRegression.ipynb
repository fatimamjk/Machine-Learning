## Now Fit The Quadratic Model

 As our Quadratic Model was
 
![Cost Funtion for quadratic ](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Cost%20funtion2.png?raw=true

And using derivatives we transformed our model into 3 simultaneous equations

NOTE :  Now you are required to compute values wihtout using foor loop on tarining data
- HINT : Use numpy library for this purpose 

```python
#Your code Here
#For loop is not recommended in these cases, 
#use numpy functions to calculate the values of above variables in one line of code.
#Such as x.shape for number of rows, sum() for sum of all elements in x
#np.dot() or multiply suntion for square of x 


#CALCULATING PREVIOUSLY CALCULATED VALUES AGAIN HERE AS IT WAS NOT GIVING AN ACCURATE ANSWER

trainData = pd.read_csv('trainRegression.csv')  #get traaing data
x_train = np.array(trainData['X']) #extract X and R col
y_train = np.array(trainData['R'])
n = x_train.shape[0]  #no. of data points
sum_x = np.sum(x_train)  # x values sum
sum_y = np.sum(y_train)  # y values sum
sum_x_square = np.sum(x_train ** 2)  # Sum of x^2 (x square)
sum_x_cube = np.sum(x_train ** 3)  # Sum of x^3    (x cube)
sum_x_fourth = np.sum(x_train ** 4)  # Sum of x^4   (x fourth)
sum_xy = np.sum(x_train * y_train)  # Sum of xy    (xy)
sum_x_square_y = np.sum((x_train ** 2) * y_train)  # Sum of x^2 * y  (x square * y )

# construct matrix A
A = np.array([[n, sum_x, sum_x_square],
              [sum_x, sum_x_square, sum_x_cube],
              [sum_x_square, sum_x_cube, sum_x_fourth]])

# Construct matrix B
B = np.array([[sum_y],
              [sum_xy],
              [sum_x_square_y]])

# Print matrices A and B
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)




```

Expected Output

![Quadratic Model Matrixes](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Matrixs%20Quad1.png?raw=true)

#### Compute Valus Of Three `s


```python
#Your code Here
#Calculate the values of 0_node and 0_1 using Matrix Multiplication
#Hint: Use X = A^-1 * B to calculate the values of 0_node , 0_1 and 0_2


# Calculate A matrix inverse
A_inverse = np.linalg.inv(A)

# Calculate the values of theta using matrix multiplication
X = np.dot(A_inverse, B)

print(" Matrix X for the quadratic Model: ")
print(X)

```
